```python
# # The dataset is highly unbalanced
# 0 --> Normal Transactions
# 1 --> Fraud Transactions


#separating the data for analysis

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
#Under sampling
```

```python
# build a sample dataset from this original dataset containing normal transactions and fraud transactions
# normal transaction is 88.291022 and fraud transaction is 122.211321
```

```python
legit_sample = legit.sample(n=492)
```

```python
#Concat two dataframes ## axis=0 means rows and axis=1 means column
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Split the data into Features(X V1 V2 etc ) and Targets(0 or 1)
#axis = 1 is column
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# #split the data into training and testing #Mention four variables
# taking the X and Y features into training and testing 
# stratify means evenly disctributed data
# random state how you want your data to be splitted
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X_train.shape)
```

```python
#Model Training
```

```python
model = LogisticRegression()
```

```python
#Training the logistic reg model with training data
```

```python
model.fit(X_train, Y_train)
```

```python
# # Model Evaluation
# Accuracy Score
```

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ',training_data_accuracy)
```

```python
#accurancy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of this data: ', test_data_accuracy)
```
